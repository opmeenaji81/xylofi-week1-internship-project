# Task 3 — Model Building

```python
# importing required libraries

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import (mean_absolute_error, mean_squared_error, r2_score)
```

```python
# spliting the dataset (80 : 20)

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size = 0.20, random_state = 42)

print("Training Data:",X_train.shape)
print("Testing Data:", X_test.shape)
```

### Linear Regression

```python
lr_model = LinearRegression()

lr_model.fit(X_train, y_train)
```

```python
lr_predictions = lr_model.predict(X_test)
```

```python
import numpy as np

lr_mse = mean_squared_error(y_test, lr_predictions)
lr_rmse = np.sqrt(lr_mse)
lr_r2 = r2_score(y_test, lr_predictions)

print("Linear Regression Performance")
print("------------------------------")
print("MAE :", lr_mae)
print("RMSE:", lr_rmse)
print("R² Score:", lr_r2)
```

### Random Forest Regressor

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
rf_predictions = rf_model.predict(X_test)
```

```python
rf_mse = mean_squared_error(y_test, rf_predictions)
rf_rmse = np.sqrt(rf_mse)
rf_r2 = r2_score(y_test, rf_predictions)

print("Random Forest Performance")
print("-------------------------")
print("MAE :", rf_mse)
print("RMSE:", rf_rmse)
print("R² Score:", rf_r2)
```

### Compare Both Models



```python
comparison = pd.DataFrame({
    "Model": [
        "Linear Regression",
        "Random Forest"
    ],
    "MAE": [
        lr_mae,
        rf_mse
    ],
    "RMSE": [
        lr_rmse,
        rf_rmse
    ],
    "R² Score": [
        lr_r2,
        rf_r2
    ]
})

comparison
    
```

1. MAE (Mean Absolute Error)
  * Measures the average prediction error.
  * Lower MAE means better performance.
2. RMSE (Root Mean Squared Error)
  * Gives more weight to larger errors.
  * Lower RMSE is better.
3. R² Score
  * Measures how well the model explains the variation in house prices.
  * Values range from 0 to 1.
  * Closer to 1 indicates a better model.

```python
# chossing best model

best_model = comparison.loc[comparison["R² Score"].idxmax(),
    "Model"
    ]

print("Best Performing Model:", best_model)
```
